# 1. IUN (Importance and Urgency) 점수 정의 및 평가:

```
from transformers import pipeline
import pandas as pd
import numpy as np


# 1. Hugging Face 텍스트 분류 모델 로드
# 논문에서 언급된 distilbart-mnli-12-1의 기반 모델 중 하나인 bart-large-mnli 사용.
# Zero-shot classification을 사용하여 민원의 중요도/긴급도와 관련된 레이블을 예측
print("Hugging Face 텍스트 분류 모델 로드 중 (bart-large-mnli)...")
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# IUN 점수 매핑을 위한 후보 레이블 정의
# 논문의 1-5점 척도에 대략적으로 매핑될 수 있도록 정의
candidate_labels = [
    "매우 중요하고 긴급함",  # 5점
    "중요하고 긴급함",      # 4점
    "보통",             # 3점
    "중요하지 않음",        # 2점
    "민원이 아님"          # 1점
]

# 각 레이블에 IUN 점수를 매핑하는 딕셔너리
label_to_iun_score = {
    "민원이 아님": 1,
    "중요하지 않음": 2,
    "보통": 3,
    "중요하고 긴급함": 4,
    "매우 중요하고 긴급함": 5
}

```

```
def get_iun_score_from_hf(text):
    """
    Hugging Face zero-shot classification을 사용하여 텍스트의 IUN 점수 (1-5)를 추론
    가장 높은 확률을 가진 레이블을 찾아 IUN 점수로 매핑
    """
    try:
        # zero-shot classification 수행
        result = classifier(text, candidate_labels, multi_label=False)

        # 가장 높은 확률을 가진 레이블 추출
        predicted_label = result['labels'][0]

        # 레이블을 IUN 점수로 매핑
        score = label_to_iun_score.get(predicted_label, 3) # 매핑되지 않는 경우 기본값 3
        return score
    except Exception as e:
        print(f"Hugging Face 추론 오류 발생: {e}. 텍스트: '{text}'")
        return 3 # 오류 발생 시 기본값 3

# 각 민원 텍스트에 IUN 점수 부여
iun_scores_hf = []
```

# 2. 뉴스 기사 클러스터링

```
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from umap import UMAP
from hdbscan import HDBSCAN
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler # KMeans 사용 시 스케일링 권장


# 1. 텍스트 임베딩 (Embedding)
# 논문에서 사용된 세 가지 임베딩 모델
embedding_models = {
    "MPN": "all-mpnet-base-v2",
    "L6": "all-MiniLM-L6-v2",
    "E5": "intfloat/multilingual-e5-small" # multilingual 모델 사용
}

embedded_data = {}

print("텍스트 임베딩 진행 중...")
for model_name, model_path in embedding_models.items():
    print(f"  - 모델: {model_name} ({model_path})")
    model = SentenceTransformer(model_path)
    embeddings = model.encode(combined_df_final, show_progress_bar=True, normalize_embeddings=True)
    embedded_data[model_name] = embeddings
    print(f"    -> 임베딩 차원: {embeddings.shape[1]}\n")

# -----------------------------------------------------------------------------
# 2. 차원 축소 (Dimension Reduction) - UMAP
# 논문 설정: D=20에서 D=3으로 차원 축소, n_neighbors=10, min_dist=0
umap_reducer = UMAP(n_neighbors=10, n_components=3, min_dist=0.0, random_state=42)

reduced_embedded_data = {}

print("차원 축소 (UMAP) 진행 중...")
for model_name, embeddings in embedded_data.items():
    print(f"  - 임베딩: {model_name}")
    reduced_embeddings = umap_reducer.fit_transform(embeddings)
    reduced_embedded_data[model_name] = reduced_embeddings
    print(f"    -> 축소된 차원: {reduced_embeddings.shape[1]}\n")

# -----------------------------------------------------------------------------
# 3. 클러스터링 (Clustering)
# 논문에서 사용된 세 가지 클러스터링 알고리즘
# 클러스터 수 범위: 20 ~ 100 (10단계 간격)
# min_cluster_size=5 (HDBSCAN)

clustering_algorithms = {
    "HDBSCAN": HDBSCAN(min_cluster_size=5, prediction_data=True), # prediction_data=True는 새로운 데이터 예측에 필요
    "KMeans": KMeans(n_clusters=30, random_state=42, n_init='auto'), # 논문 범위 내에서 임의의 클러스터 수 설정
    "Agglomerative": AgglomerativeClustering(n_clusters=30, linkage='ward') # 논문과 동일하게 'ward' linkage 사용
}

# 클러스터링 결과 저장 딕셔너리
clustering_results = {}

print("클러스터링 진행 중...")
for embed_model_name, reduced_embeddings in reduced_embedded_data.items():
    clustering_results[embed_model_name] = {}
    print(f"  - 임베딩: {embed_model_name}")

    # KMeans는 스케일링된 데이터에 더 잘 작동할 수 있습니다.
    scaler = StandardScaler()
    scaled_embeddings = scaler.fit_transform(reduced_embeddings)

    for algo_name, algo_model in clustering_algorithms.items():
        print(f"    - 알고리즘: {algo_name}")

        # KMeans는 스케일링된 임베딩 사용
        if algo_name == "KMeans":
            clusters = algo_model.fit_predict(scaled_embeddings)
        else:
            clusters = algo_model.fit_predict(reduced_embeddings)

        clustering_results[embed_model_name][algo_name] = clusters
        unique_clusters = np.unique(clusters)
        print(f"      -> 발견된 클러스터 수: {len(unique_clusters)} (노이즈 포함: {(-1 in unique_clusters)})")

        # 각 클러스터의 데이터 포인트 수 확인 (min_cluster_size=3 조건 확인)
        for cluster_id in unique_clusters:
            count = np.sum(clusters == cluster_id)
            if cluster_id != -1 and count < 3:
                print(f"        주의: 클러스터 {cluster_id}에 3개 미만의 데이터 포인트({count}개)가 있습니다.")

        print()

# -----------------------------------------------------------------------------
# 4. 결과 확인 (예시)
print("--- 클러스터링 결과 요약 ---")
for embed_model_name, algo_results in clustering_results.items():
    print(f"\n[임베딩 모델: {embed_model_name}]")
    for algo_name, clusters in algo_results.items():
        print(f"  - 클러스터링 알고리즘: {algo_name}")
        # 예시로 첫 10개 민원의 클러스터 ID 출력
        print(f"    첫 10개 민원의 클러스터 ID: {clusters[:10]}")
        # 각 클러스터에 속하는 데이터 포인트 수
        unique_clusters, counts = np.unique(clusters, return_counts=True)
        cluster_summary = dict(zip(unique_clusters, counts))
        print(f"    클러스터 분포: {cluster_summary}")

print("\n클러스터링 프로세스 완료.")
print("이제 이 클러스터링 결과를 바탕으로 논문의 4단계 (D90, D50 계산 및 IUN 스코어와의 상관관계 분석)를 진행할 수 있습니다.")
```

# 3. 클러스터별 IUN 스코어 할당

```
# 1. 클러스터별 IUN 스코어 평균 할당
# 결과를 저장할 딕셔너리
cluster_iun_scores = {}

# IUN 점수 컬럼 이름 지정
IUN_SCORE_COLUMN_NAME = 'IUN 점수 (HF)'

print("클러스터별 IUN 스코어 평균 계산 중...")
for embed_model_name, algo_results in clustering_results.items():
    cluster_iun_scores[embed_model_name] = {}
    print(f"\n[임베딩 모델: {embed_model_name}]")

    for algo_name, clusters in algo_results.items():
        print(f"  - 클러스터링 알고리즘: {algo_name}")

        # --- 클러스터 배열과 df_iun_hf 길이 일치 확인---
        if len(clusters) != len(data):
            print(f"    경고: 클러스터 배열 '{algo_name}'의 길이가 df_iun_hf({len(data)})와 일치하지 않습니다({len(clusters)}). 이 클러스터링 결과는 건너뜁니다.")
            continue

        # 유니크한 클러스터 ID들을 가져옴 (노이즈 클러스터 -1 포함)
        unique_cluster_ids = np.unique(clusters)

        # 각 클러스터의 IUN 점수 평균을 저장할 딕셔너리
        current_algo_cluster_iun = {}

        for cluster_id in unique_cluster_ids:
            # 노이즈 클러스터 (-1)는 일반적으로 평균 계산에서 제외
            if cluster_id == -1:
                print(f"    - 클러스터 {cluster_id} (노이즈): IUN 점수 계산에서 제외.")
                continue

            # 현재 클러스터 ID에 속하는 데이터 포인트들의 인덱스를 찾음
            indices_in_cluster = np.where(clusters == cluster_id)[0]

            # --- 수정된 부분: df_iun_hf DataFrame에서 IUN 점수들을 가져옴 ---
            if IUN_SCORE_COLUMN_NAME: # 컬럼명이 지정된 DataFrame인 경우
                iun_scores_in_cluster = data.loc[indices_in_cluster, IUN_SCORE_COLUMN_NAME].tolist()
            else: # df_iun_hf가 직접 Series인 경우 (또는 단일 컬럼 DataFrame)
                # 이 경우, df_iun_hf는 길이가 clusters와 같은 Series여야 합니다.
                iun_scores_in_cluster = data.iloc[indices_in_cluster].tolist()
                print(f"    - 주의: '{IUN_SCORE_COLUMN_NAME}' 컬럼명이 지정되지 않아 data 전체를 인덱싱하여 점수를 가져옵니다. df_iun_hf가 Series 형태인지 확인해주세요.")


            # 클러스터 내 IUN 점수들의 평균을 계산.
            if iun_scores_in_cluster: # 빈 클러스터가 아닌 경우에만 계산
                avg_iun = np.mean(iun_scores_in_cluster)
            else:
                avg_iun = 0 # 클러스터가 비어있으면 0으로 설정

            current_algo_cluster_iun[cluster_id] = avg_iun

        cluster_iun_scores[embed_model_name][algo_name] = current_algo_cluster_iun

        # 계산된 클러스터별 평균 IUN 점수 출력
        print(f"    -> 클러스터별 평균 IUN 점수: {current_algo_cluster_iun}")

print("\n--- 클러스터별 IUN 스코어 평균 할당 완료 ---")

```

# 4. 클러스터의 기하학적 속성(D90 및 D50) 계산

```
import numpy as np
import pandas as pd
from scipy.spatial.distance import cdist # 거리 계산을 위해 필요


# 1. 클러스터의 기하학적 속성 ($D_{90}$, $D_{50}$, $D_{90}-D_{50}$) 계산
geometric_properties = {}

print("클러스터의 기하학적 속성 ($D_{90}$, $D_{50}$) 계산 중...")
for embed_model_name, embeddings in reduced_embedded_data.items():
    geometric_properties[embed_model_name] = {}
    print(f"\n[임베딩 모델: {embed_model_name}]")

    for algo_name, clusters in clustering_results[embed_model_name].items():
        geometric_properties[embed_model_name][algo_name] = {}
        print(f"  - 클러스터링 알고리즘: {algo_name}")

        unique_cluster_ids = np.unique(clusters)

        for cluster_id in unique_cluster_ids:
            # 노이즈 클러스터 (-1)는 제외
            if cluster_id == -1:
                continue

            # 해당 클러스터에 속하는 데이터 포인트들의 인덱스
            indices_in_cluster = np.where(clusters == cluster_id)[0]

            # 클러스터에 데이터 포인트가 없으면 건너뜀
            if len(indices_in_cluster) == 0:
                continue

            # -----------------------------------------------------------------
            # 1.1. 클러스터의 중심 (centroid) 계산
            cluster_points = embeddings[indices_in_cluster]
            cluster_center = np.mean(cluster_points, axis=0)

            # -----------------------------------------------------------------
            # 1.2. 각 클러스터 포인트와 중심 간의 거리 계산
            # cdist는 두 점 집합 간의 모든 거리 쌍을 계산
            # 여기서는 (N, D) 형태의 cluster_points와 (1, D) 형태의 cluster_center 간의 거리를 계산
            distances_to_center = cdist(cluster_points, cluster_center.reshape(1, -1), metric='euclidean').flatten()

            # -----------------------------------------------------------------
            # 1.3. $D_{90}$ 및 $D_{50}$ 계산
            # np.percentile을 사용하여 백분위수 거리(반경)
            d90 = np.percentile(distances_to_center, 90)
            d50 = np.percentile(distances_to_center, 50)
            d90_minus_d50 = d90 - d50

            geometric_properties[embed_model_name][algo_name][cluster_id] = {
                'D90': d90,
                'D50': d50,
                'D90_minus_D50': d90_minus_d50
            }

            print(f"    -> 클러스터 {cluster_id}: D90={d90:.3f}, D50={d50:.3f}, D90-D50={d90_minus_d50:.3f}")

print("\n--- 클러스터의 기하학적 속성 계산 완료 ---")

# -----------------------------------------------------------------------------
# 2. 결과 확인
print("\n--- 최종 클러스터 기하학적 속성 요약 ---")
for embed_model_name, algo_results in geometric_properties.items():
    print(f"\n[임베딩 모델: {embed_model_name}]")
    for algo_name, cluster_props in algo_results.items():
        print(f"  - 클러스터링 알고리즘: {algo_name}")

        # 결과를 DataFrame으로 변환하여 보기 좋게 출력
        df_props = pd.DataFrame.from_dict(cluster_props, orient='index')
        df_props.index.name = 'Cluster ID'
        print(df_props.to_string())
        print()
```

```

# --- 새로운 민원 처리 함수 ---

def process_new_complaint(new_complaint_text, embedding_models, umap_reducer, clustering_algorithms,
                          embedded_data, reduced_embedded_data, clustering_results, cluster_iun_scores, geometric_properties,
                          classifier, candidate_labels, label_to_iun_score, IUN_SCORE_COLUMN_NAME='IUN 점수 (HF)'):
    """
    새로운 민원 텍스트를 처리하여 IUN 스코어를 부여하고,
    기존 클러스터에 할당하며, 긴급도를 산정합니다.
    """
    print(f"\n--- 새로운 민원 처리 시작: '{new_complaint_text[:50]}...' ---")

    # 1. 새로운 민원에 IUN 스코어 부여
    new_iun_score = get_iun_score_from_hf(new_complaint_text)
    print(f"새로운 민원의 개별 IUN 스코어: {new_iun_score}")

    # 2. 새로운 민원 임베딩 및 차원 축소
    new_embeddings = {}
    new_reduced_embeddings = {}
    for model_name, model_path in embedding_models.items():
        model = SentenceTransformer(model_path)
        new_embedding = model.encode([new_complaint_text], normalize_embeddings=True)
        new_embeddings[model_name] = new_embedding[0] # 단일 임베딩이므로 [0]으로 추출

        # UMAP을 사용하여 차원 축소 (기존 fit된 reducer 사용)
        # UMAP은 transform만으로도 새로운 데이터 처리 가능
        new_reduced_embedding = umap_reducer.transform(new_embedding)
        new_reduced_embeddings[model_name] = new_reduced_embedding[0] # 단일 임베딩이므로 [0]으로 추출
        print(f"  - {model_name} 임베딩 및 차원 축소 완료.")

    # 3. 기존 클러스터 중 가장 유사한 클러스터에 할당
    assigned_clusters = {}
    predicted_cluster_details = {}

    print("\n기존 클러스터에 할당 중...")
    for embed_model_name, new_reduced_embed in new_reduced_embeddings.items():
        predicted_cluster_details[embed_model_name] = {}
        print(f"  [임베딩 모델: {embed_model_name}]")

        for algo_name, algo_model in clustering_algorithms.items():
            print(f"    - 클러스터링 알고리즘: {algo_name}")

            assigned_cluster_id = -1 # 기본값은 할당 실패 또는 노이즈

            if algo_name == "HDBSCAN":
                # HDBSCAN은 approximate_predict를 사용하여 새로운 데이터 예측
                # prediction_data=True로 초기화된 모델만 가능
                try:
                    # 'relative_membership_vectors' 또는 'labels'를 반환
                    # 여기서는 'labels'를 사용하여 가장 가능성 높은 클러스터 ID를 얻음
                    if hasattr(algo_model, 'prediction_data') and algo_model.prediction_data:

                        # 1. 클러스터별 평균 임베딩 계산 (이전 코드에서 계산되지 않았다면 여기서 계산)
                        cluster_centers = {}
                        for cluster_id in np.unique(clustering_results[embed_model_name][algo_name]):
                            if cluster_id == -1: continue # 노이즈 클러스터 제외

                            indices = np.where(clustering_results[embed_model_name][algo_name] == cluster_id)[0]
                            # 주의: UMAP 축소된 임베딩을 사용해야 합니다.
                            cluster_centers[cluster_id] = np.mean(reduced_embedded_data[embed_model_name][indices], axis=0)

                        if cluster_centers: # 클러스터가 존재하는 경우에만 처리
                            # 새로운 임베딩과 각 클러스터 중심 간의 거리 계산
                            distances = {
                                c_id: np.linalg.norm(new_reduced_embed - center)
                                for c_id, center in cluster_centers.items()
                            }
                            # 가장 가까운 클러스터 ID 선택
                            assigned_cluster_id = min(distances, key=distances.get)
                            # 소속도 개념은 HDBSCAN의 predict_proba와 더 관련 깊지만, 여기서는 간략하게 처리
                            print(f"      -> HDBSCAN: 가장 가까운 클러스터 {assigned_cluster_id}에 할당 (거리: {distances[assigned_cluster_id]:.3f})")
                        else:
                            print("      -> HDBSCAN: 클러스터 중심을 찾을 수 없어 할당 실패.")
                            assigned_cluster_id = -1 # 클러스터 할당 실패
                    else:
                        print("      -> HDBSCAN: 'prediction_data=True'로 초기화되지 않아 예측을 건너뜀. 직접 구현 필요.")
                        assigned_cluster_id = -1 # 예측 불가
                except Exception as e:
                    print(f"      -> HDBSCAN 예측 중 오류 발생: {e}. 할당 실패.")
                    assigned_cluster_id = -1

            elif algo_name == "KMeans":
                # KMeans는 새로운 데이터에 대해 predict 메소드를 제공
                # 새로운 임베딩에 대한 predict는 2D array 형태를 요구
                assigned_cluster_id = algo_model.predict(new_reduced_embed.reshape(1, -1))[0]
                # 할당된 클러스터의 중심과의 거리도 계산 가능
                distance_to_center = np.linalg.norm(new_reduced_embed - algo_model.cluster_centers_[assigned_cluster_id])
                print(f"      -> KMeans: 클러스터 {assigned_cluster_id}에 할당 (중심까지의 거리: {distance_to_center:.3f})")

            elif algo_name == "Agglomerative":
                # AgglomerativeClustering은 predict 메소드를 제공하지 않음
                # 가장 가까운 기존 클러스터의 중심을 찾아 할당
                # 1. 클러스터별 평균 임베딩 계산
                cluster_centers = {}
                for cluster_id in np.unique(clustering_results[embed_model_name][algo_name]):
                    indices = np.where(clustering_results[embed_model_name][algo_name] == cluster_id)[0]
                    # 주의: UMAP 축소된 임베딩을 사용해야 함
                    cluster_centers[cluster_id] = np.mean(reduced_embedded_data[embed_model_name][indices], axis=0)

                if cluster_centers:
                    distances = {
                        c_id: np.linalg.norm(new_reduced_embed - center)
                        for c_id, center in cluster_centers.items()
                    }
                    assigned_cluster_id = min(distances, key=distances.get)
                    print(f"      -> Agglomerative: 가장 가까운 클러스터 {assigned_cluster_id}에 할당 (거리: {distances[assigned_cluster_id]:.3f})")
                else:
                    print("      -> Agglomerative: 클러스터 중심을 찾을 수 없어 할당 실패.")
                    assigned_cluster_id = -1

            assigned_clusters[f"{embed_model_name}_{algo_name}"] = assigned_cluster_id

            # 할당된 클러스터의 평균 IUN 스코어 및 D90-D50 값 가져오기
            cluster_avg_iun = cluster_iun_scores.get(embed_model_name, {}).get(algo_name, {}).get(assigned_cluster_id, 'N/A')
            cluster_d90_d50 = geometric_properties.get(embed_model_name, {}).get(algo_name, {}).get(assigned_cluster_id, {}).get('D90_minus_D50', 'N/A')

            predicted_cluster_details[embed_model_name][algo_name] = {
                'assigned_cluster_id': assigned_cluster_id,
                'cluster_avg_iun_score': cluster_avg_iun,
                'cluster_D90_minus_D50': cluster_d90_d50
            }
            print(f"      -> 할당된 클러스터 {assigned_cluster_id}의 평균 IUN 스코어: {cluster_avg_iun}")
            print(f"      -> 할당된 클러스터 {assigned_cluster_id}의 D90-D50: {cluster_d90_d50:.3f}" if isinstance(cluster_d90_d50, (int, float)) else f"      -> 할당된 클러스터 {assigned_cluster_id}의 D90-D50: {cluster_d90_d50}")


    # 4. 민원 긴급도 산정 (개념적 설명)
    print("\n--- 민원 긴급도 산정 ---")
    print("논문에 따르면, 새로운 민원의 긴급도는 다음 세 가지 요소를 종합하여 산정됩니다:")
    print("1. **개별 IUN 스코어**: Hugging Face 모델이 새로운 민원에 직접 부여한 IUN 스코어.")
    print("2. **클러스터 평균 IUN 스코어**: 할당된 클러스터의 평균 IUN 스코어.")
    print("3. **클러스터의 기하학적 특성 (D90-D50)**: 할당된 클러스터의 D90-D50 값.")
    print("\n이 세 가지 값을 어떻게 가중치 부여하여 최종 긴급도를 산정할지는 추가적인 분석 및 정책 결정이 필요합니다.")
    print("예를 들어, 다음과 같은 방식으로 가중치를 부여할 수 있습니다:")
    print("Final_Urgency_Score = (Weight1 * new_iun_score) + (Weight2 * cluster_avg_iun_score) + (Weight3 * (1 / cluster_D90_minus_D50))")
    print("(D90-D50 값이 작을수록 클러스터가 밀집되어 중요도가 높을 수 있으므로 역수를 취하거나 음의 가중치를 부여할 수 있습니다.)")
    print("현재 계산된 값들은 다음과 같습니다:")
    print(f" - 새로운 민원 개별 IUN 스코어: {new_iun_score}")
    for embed_model_name, algo_results in predicted_cluster_details.items():
        for algo_name, details in algo_results.items():
            print(f" - 할당된 클러스터 정보 ({embed_model_name}, {algo_name}):")
            print(f"    - 클러스터 ID: {details['assigned_cluster_id']}")
            print(f"    - 클러스터 평균 IUN 스코어: {details['cluster_avg_iun_score']}")
            print(f"    - 클러스터 D90-D50: {details['cluster_D90_minus_D50']:.3f}" if isinstance(details['cluster_D90_minus_D50'], (int, float)) else f"    - 클러스터 D90-D50: {details['cluster_D90_minus_D50']}")


    print("\n--- 새로운 민원 처리 완료 ---")
    return {
        'new_iun_score': new_iun_score,
        'assigned_clusters': assigned_clusters,
        'predicted_cluster_details': predicted_cluster_details
    }
```

# 5. IUN 스코어와 기하학적 속성 간의 상관관계 분석

```
# -----------------------------------------------------------------------------
# 5. IUN 스코어와 기하학적 속성 간의 켄달 타우 상관관계 분석
correlation_results = {}

print("\nIUN 스코어와 기하학적 속성 간의 켄달 타우 상관관계 계산 중...")
for embed_model_name in cluster_iun_scores.keys():
    correlation_results[embed_model_name] = {}
    print(f"\n[임베딩 모델: {embed_model_name}]")

    for algo_name in cluster_iun_scores[embed_model_name].keys():
        correlation_results[embed_model_name][algo_name] = {}
        print(f"  - 클러스터링 알고리즘: {algo_name}")

        # 해당 조합의 클러스터 ID, 평균 IUN 스코어, 기하학적 속성 데이터를 추출
        cluster_ids_to_analyze = sorted(cluster_iun_scores[embed_model_name][algo_name].keys())

        if not cluster_ids_to_analyze:
            print("    -> 분석할 클러스터가 없습니다. (노이즈 제외 후)")
            continue

        avg_iun_values = [cluster_iun_scores[embed_model_name][algo_name][cid] for cid in cluster_ids_to_analyze]
        d90_values = [geometric_properties[embed_model_name][algo_name][cid]['D90'] for cid in cluster_ids_to_analyze]
        d50_values = [geometric_properties[embed_model_name][algo_name][cid]['D50'] for cid in cluster_ids_to_analyze]
        d90_minus_d50_values = [geometric_properties[embed_model_name][algo_name][cid]['D90_minus_D50'] for cid in cluster_ids_to_analyze]

        # -----------------------------------------------------------------
        # 켄달 타우 상관관계 계산
        # Kendall's tau correlation coefficient (타우, p-value)

        # IUN 스코어와 D90 간의 상관관계
        tau_d90, p_d90 = kendalltau(avg_iun_values, d90_values)
        correlation_results[embed_model_name][algo_name]['IUN_vs_D90'] = {'tau': tau_d90, 'p_value': p_d90}
        print(f"    - IUN vs D90 (tau): {tau_d90:.4f} (p={p_d90:.4f})")

        # IUN 스코어와 D50 간의 상관관계
        tau_d50, p_d50 = kendalltau(avg_iun_values, d50_values)
        correlation_results[embed_model_name][algo_name]['IUN_vs_D50'] = {'tau': tau_d50, 'p_value': p_d50}
        print(f"    - IUN vs D50 (tau): {tau_d50:.4f} (p={p_d50:.4f})")

        # IUN 스코어와 D90 - D50 간의 상관관계
        tau_d90_d50_diff, p_d90_d50_diff = kendalltau(avg_iun_values, d90_minus_d50_values)
        correlation_results[embed_model_name][algo_name]['IUN_vs_D90_minus_D50'] = {'tau': tau_d90_d50_diff, 'p_value': p_d90_d50_diff}
        print(f"    - IUN vs (D90-D50) (tau): {tau_d90_d50_diff:.4f} (p={p_d90_d50_diff:.4f})")
        print()

print("\n--- 상관관계 분석 완료 ---")

# -----------------------------------------------------------------------------
# 6. 결과 요약 및 표 형태로 출력 (논문 '표 5'와 유사하게)
# 논문 표 5는 단순히 평균값을 보여주므로, 여기서는 모든 조합에 대한 결과를 보여줌
# 논문처럼 특정 조건(예: MPN, UMAP D=20)에 대한 평균을 내려면 추가적인 필터링이 필요

# 논문 '표 5'와 유사한 형태로 도출하기 위한 추가 작업
# 논문 표 5의 형식:
#             correlation with IUN    avg    stdev    Fpos
#                       D90        0.07    0.15    0.7523
#                       -D50       0.19    0.17    0.8528
#                       D90 - D50  0.35    0.13    1.0000

# Fpos는 양의 상관관계를 보이는 비율을 의미. (tau > 0)

summary_data_for_paper_table = []
all_tau_d90 = []
all_tau_d50 = []
all_tau_d90_minus_d50 = []

for embed_model_name, algo_results in correlation_results.items():
    for algo_name, corr_data in algo_results.items():
        all_tau_d90.append(corr_data['IUN_vs_D90']['tau'])
        all_tau_d50.append(corr_data['IUN_vs_D50']['tau'])
        all_tau_d90_minus_d50.append(corr_data['IUN_vs_D90_minus_D50']['tau'])

# D90에 대한 통계
avg_d90 = np.mean(all_tau_d90)
std_d90 = np.std(all_tau_d90)
fpos_d90 = np.sum(np.array(all_tau_d90) > 0) / len(all_tau_d90) if len(all_tau_d90) > 0 else 0

# D50에 대한 통계 (논문 표에는 -D50로 표기)
avg_d50 = np.mean(all_tau_d50)
std_d50 = np.std(all_tau_d50)
fpos_d50 = np.sum(np.array(all_tau_d50) > 0) / len(all_tau_d50) if len(all_tau_d50) > 0 else 0

# D90 - D50에 대한 통계
avg_d90_minus_d50 = np.mean(all_tau_d90_minus_d50)
std_d90_minus_d50 = np.std(all_tau_d90_minus_d50)
fpos_d90_minus_d50 = np.sum(np.array(all_tau_d90_minus_d50) > 0) / len(all_tau_d90_minus_d50) if len(all_tau_d90_minus_d50) > 0 else 0


summary_table_for_paper = pd.DataFrame({
    'correlation with IUN': ['D90', '-D50', 'D90 - D50'],
    'avg': [avg_d90, avg_d50, avg_d90_minus_d50],
    'stdev': [std_d90, std_d50, std_d90_minus_d50],
    'F_pos': [fpos_d90, fpos_d50, fpos_d90_minus_d50]
})

# 소수점 자리 수 조정
summary_table_for_paper['avg'] = summary_table_for_paper['avg'].round(4)
summary_table_for_paper['stdev'] = summary_table_for_paper['stdev'].round(4)
summary_table_for_paper['F_pos'] = summary_table_for_paper['F_pos'].round(4)


print("\n--- 논문 '표 5'와 유사한 최종 켄달 타우 상관관계 요약 ---")
print(summary_table_for_paper.to_string(index=False))

```
